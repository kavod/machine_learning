```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
house_raw_data = pd.read_csv('house.csv')
house_data = house_raw_data[house_raw_data['loyer']<7000]
```

```python
plt.plot(house_data['surface'],house_data['loyer'],'ro',markersize=4)
```

```python
ones = np.ones(house_data.shape[0])
x = np.matrix([ones,house_data['surface']]).T
y = house_data['loyer'].T
```

```python
theta = np.linalg.inv(x.T.dot(x)).dot(x.T).dot(y)
```

```python
theta
```

```python
100 * theta.item(1) + theta.item(0)
```

```python
plt.plot([0,250],[theta.item(0),250 * theta.item(1) + theta.item(0)])
plt.plot(house_data['surface'],house_data['loyer'],'ro',markersize=4)
plt.show()
```
